```python
df = pd.DataFrame({
    'tags': [['0119,0149,4773,4776'], ['0119,4621'], ['0142,4941'], ['0146']],
    'codes': ['10', '12', '14', '15']
})

# Define a function to extract the first two digits of each tag
def get_prefix(tag):
    return [str(t)[:2] for t in tag]

# Apply the function to the 'tags' column to create a new column with the prefixes
df['prefixes'] = df['tags'].apply(get_prefixes)

# Use the explode function to expand the 'prefixes' column into a new row for each prefix
df = df.explode('prefixes')

# Group by the 'codes' and 'prefixes' columns and count the number of occurrences
df = df.groupby(['codes', 'prefixes']).size().reset_index(name='strength')

# Display the resulting dataframe
print(df)
```

```python
def get_prefixes(codes):
    prefixes = []
    for code_list in codes:
            codes = code_list.split(',')
            for code in codes:
                if len(code) >= 2:
                    prefixes.append(code[:2])
    return prefixes
```

```python
df4['prefixes'] = df4['ADN_SECTOR_SEC_N_COD'].apply(lambda x: [str(code)[:2] for code in str(x).split(",")])

# create a list of unique prefixes
unique_prefixes = list(set([prefix for prefixes in df4['prefixes'] for prefix in prefixes]))

# create a dictionary to store the count of connections between each code and prefix
code_prefix_counts = {code: {prefix: 0 for prefix in unique_prefixes} for code in df4['CCNAE_COD']}

# update the dictionary with the counts of connections between each code and prefix
for index, row in df4.iterrows():
    for code in row['prefixes']:
        code_prefix_counts[row['CCNAE_COD']][code] += 1

# create a new dataframe to store the connections between each code and prefix
connections_df = pd.DataFrame(columns=['code', 'prefix', 'strength'])
for code, prefix_counts in code_prefix_counts.items():
    for prefix, count in prefix_counts.items():
        if count > 0:
            connections_df = connections_df.append({'code': code, 'prefix': prefix, 'strength': count}, ignore_index=True)

# print the connections dataframe
print(connections_df.head())
```
